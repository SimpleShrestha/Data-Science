```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
a=np.arange(25)
```

```python
b=a**2
```

```python
plt.plot(a,b,color='g',marker='*')
plt.show()
```

```python
plt.xlabel('Days')
plt.ylabel('Corona Cases')
plt.title('Days VS Corona')
plt.show()
```

```python
a=np.arange(20)
b=a**2
c=a**3
plt.plot(a,b,color='b')
plt.plot(a,c,color='y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X vs Y axis')
plt.show()
```
